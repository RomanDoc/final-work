```python
import pandas as pd
```

```python
df = pd.read_csv('df_prep_3_v2.csv', index_col=0)
```

```python
df
```

```python
df = df.drop('device_screen_resolution', axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, confusion_matrix

from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Нормализация датасета
list_columns = df.columns
list_columns
```

```python
scaler = StandardScaler()
```

```python
df[list_columns] = scaler.fit_transform(df)
df
```

```python
x = df.drop(['target_column'], axis=1)
y = df.target_column
```

```python
# обучение дерево решений
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier(random_state=42, class_weight='balanced', max_leaf_nodes=70)
clf.fit(x_train, y_train)
```

```python
predicted_train = clf.predict(x_train) 
predicted_test = clf.predict(x_test)
```

```python
print(roc_auc_score(y_train, predicted_train))
print(roc_auc_score(y_test, predicted_test))
```

```python
print(roc_auc_score(y_train, predicted_train))
print(roc_auc_score(y_test, predicted_test))
```

```python
confusion_matrix(y_train, predicted_train)
```

```python
confusion_matrix(y_test, predicted_test)
```

```python
f_imp_list = list(zip(x_train.columns, clf.feature_importances_))
f_imp_list.sort(key=lambda x: x[1], reverse=True)
f_imp_list
```

```python
# обучение случайный лес
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
rt_clf = RandomForestClassifier(random_state=42, max_depth=50, max_leaf_nodes=500, class_weight='balanced')
rt_clf.fit(x_train, y_train)
```

```python
predicted_train = rt_clf.predict(x_train) 
predicted_test = rt_clf.predict(x_test)
```

```python
print(roc_auc_score(y_train, predicted_train))
print(roc_auc_score(y_test, predicted_test))
```

```python
confusion_matrix(y_train, predicted_train)
```

```python
f_imp_list = list(zip(x_train.columns, rt_clf.feature_importances_))
f_imp_list.sort(key=lambda x: x[1], reverse=True)
f_imp_list
```
