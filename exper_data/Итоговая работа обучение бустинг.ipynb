```python
import pandas as pd
```

```python
df = pd.read_csv('df_prep_3_v2.csv', index_col=0)
```

```python
df
```

```python
df = df.drop('device_screen_resolution', axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, confusion_matrix

from sklearn.ensemble import GradientBoostingClassifier
```

```python
X = df.drop('target_column', axis=1)
y = df.target_column
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = GradientBoostingClassifier(random_state=42, max_leaf_nodes=70, n_estimators=500, max_depth=5, max_features=10, learning_rate=0.5)
clf.fit(X_train, y_train)
```

```python
predicted_train = clf.predict(X_train) 
predicted_test = clf.predict(x_test)
```

```python
print(roc_auc_score(y_train, predicted_train))
print(roc_auc_score(y_test, predicted_test))
```

```python
f_imp_list = list(zip(X_train.columns, clf.feature_importances_))
f_imp_list.sort(key=lambda x: x[1], reverse=True)
f_imp_list
```
